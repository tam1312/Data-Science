```python
parsed_products = []
for product in products:
  title = product.select_one('.wd-entities-title').text.strip()

  regular_price = product.select_one('del .woocommerce-Price-amount')
  if regular_price:
    regular_price = regular_price.text.strip()
  else:
    regular_price = product.select_one('.woocommerce-Price-amount')
    if regular_price:
        regular_price = regular_price.text.strip()

  discount_price = product.select_one('ins .woocommerce-Price-amount')
  if discount_price:
    discount_price = discount_price.text.strip()
  else:
    discount_price = None

  product_page = product.select_one('.wd-entities-title a')
  if product_page:
    product_page = product_page.get('href')

  add_to_cart_button = product.select_one('.wd-add-btn a')
  if add_to_cart_button:
    add_to_cart_button = product_page + add_to_cart_button.get('href')

  product_dict = {
      "Product Title": title,
      "Regular Price": regular_price,
      "Discount Price": discount_price,
      "Product Page": product_page,
      "Button URL": add_to_cart_button
  }

  parsed_products.append(product_dict)
```

## Repeat the extraction process for each page of products

```python
base_url = 'https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/'
```

```python
def extract_product_characteristics_as_dict(product):
  title = product.select_one('.wd-entities-title').text.strip()

  regular_price = product.select_one('del .woocommerce-Price-amount')
  if regular_price:
    regular_price = regular_price.text.strip()
  else:
    regular_price = product.select_one('.woocommerce-Price-amount')
    if regular_price:
        regular_price = regular_price.text.strip()

  discount_price = product.select_one('ins .woocommerce-Price-amount')
  if discount_price:
    discount_price = discount_price.text.strip()
  else:
    discount_price = None

  product_page = product.select_one('.wd-entities-title a')
  if product_page:
    product_page = product_page.get('href')

  add_to_cart_button = product.select_one('.wd-add-btn a')
  if add_to_cart_button:
    add_to_cart_button = product_page + add_to_cart_button.get('href')

  product_dict = {
      "Product Title": title,
      "Regular Price": regular_price,
      "Discount Price": discount_price,
      "Product Page": product_page,
      "Button URL": add_to_cart_button
  }

  return product_dict
```

```python
all_products = []

for i in range(1,14):
  url = base_url + str(i)
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  products = soup.select('.products .product')

  for product in products:
    result = extract_product_characteristics_as_dict(product)
    all_products.append(result)
```
